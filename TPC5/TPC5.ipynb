```python
#Modelo
#Cinema = [Sala]
#Sala = [NumLugares, Vendidos, Filme, Sala]
#NumLugares = Int
#Filme = String
#Vendidos = [Int]

def inserirSala(cinema, sala):
    cinema.append(sala)
    return cinema

def listar(cinema):
    print("-----------------------Filmes em Exibição-----------------------")
    for sala in cinema:
        print(sala[3]," ",sala[2]," | ", "Nº de lugares: ",sala[0])

def disponivel(cinema, filme, lugar):
    cond = False
    for sala in cinema:
        if sala[2] == filme:
            if lugar not in sala[1]:
                cond = True
    return cond

def vendeBilhete(cinema, filme, lugar):
    if disponivel(cinema, filme, lugar):
        for sala in cinema:
            if sala[2] == filme:
                sala[1].append(lugar)
    return cinema

def listarDisponibilidades(cinema):
    print("-----------------------Filmes em Exibição-----------------------")
    for sala in cinema:
        nLugaresDisponiveis = sala[0] - len(sala[1])
        print(sala[3]," ",sala[2]," | ", "Nº de Lugares Disponíveis: ",nLugaresDisponiveis)

def listarSala(cinema, filme):
    for sala in cinema:
        if sala[2] == filme:
            print("-----------------------Filme", sala[2],"-----------------------")
            print(sala[2]," | Nº de Lugares:",sala[0]," | Nº de Lugares ocupados:", sala[1])

def devolveBilhete(cinema, filme, lugar):
    for sala in cinema:
        if sala[2] == filme:
            sala[1].remove(lugar)
    return cinema

def criarSala(cinema, filme2):
    numLugares = int(input("Insira o número de lugares da sua sala:"))
    vendidos = []
    sala2 = input("Introduza o número da sala que prentende criar, no formato 'Sala X':")
    sala = (numLugares, vendidos, filme2, sala2)
    cinema.append(sala)
    return cinema

def removerSala(cinema, filme):
    for sala in cinema:
        if sala[2] == filme:
            if sala[1] == []:
                cinema.remove(sala)

def menu():
    print(""" 
    (1) Criar sala
    (2) Remover sala
    (3) Listar sala
    (4) Consultar cartaz em exibição
    (5) Verificar a disponibilidade relativamente à capacidade máxima da sala
    (6) Verificar se x lugar está disponível
    (7) Comprar bilhete
    (8) Devolver bilhete
    (0) Sair
    """)
```

```python
#Teste às funções

sala1 = (300, [], "Oppenheimer", "Sala 1")
sala2 = (350, [], "Barbie", "Sala 2")
sala3 = (250, [], "Patrulha Pata: o Super Filme", "Sala 3")

cinema = []
cinema = inserirSala(cinema, sala1)
cinema = inserirSala(cinema, sala2)
cinema = inserirSala(cinema, sala3)

listar(cinema)

print(disponivel(cinema, "Oppenheirmer", 30))
print(disponivel(cinema, "Barbie", 59))
print(disponivel(cinema, "Patrulha Pata: o Super Filme", 85))

for j in range(1,130):
    cinema = vendeBilhete(cinema, "Oppenheirmer", j + 50)
    cinema = vendeBilhete(cinema, "Barbie", j)
    cinema = vendeBilhete(cinema, "Patrulha Pata: o Super Filme", j*2)

listarDisponibilidades(cinema)

listar(cinema)

print(cinema[1])
```

```python
#Aplicação para gerir cinema

menu()
n = int(input("Introduza uma funcionalidade do menu:"))

while n != 0:
    if n == 1:
        filme2 = input("Introduza o nome do filme que pretende adicionar ao cartaz:")
        criarSala(cinema, filme2)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 2:
        filme = input("Insira o nome do filme exibido na sala que pretende remover:")
        removerSala(cinema, filme)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 3:
        filme = input("Insira o nome do filme em exibição que pretende listar:")
        listarSala(cinema, filme)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 4:
        listar(cinema)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 5:
        listarDisponibilidades(cinema)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 6:
        filme = input("Insira o nome do filme que pretende assistir")
        lugar = int(input("De acordo com a capacidade da sala escolhida, escolha o seu lugar:"))
        disponivel(cinema, filme, lugar)
        print(disponivel(cinema, filme, lugar))
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 7:
        filme = input("Insira o nome do filme que pretende assistir")
        lugar = int(input("Escolha o seu lugar:"))
        vendeBilhete(cinema, filme, lugar)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
    elif n == 8:
        filme = input("Insira o nome do filme do bilhete que pretende devolver")
        lugar = int(input("Insira qual era o seu lugar:"))
        devolveBilhete(cinema, filme, lugar)
        n = int(input("Introduza a próxima funcionalidade do menu que pretende aceder:"))
```
